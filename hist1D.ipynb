```python
from info_file import *
import matplotlib.pyplot as plt
import glob
import fileinput
```

```python
file_list = glob.glob("run_1M_09_02_2023\*.txt")

with open('result.txt', 'w') as file:
    input_lines = fileinput.input(file_list)
    file.writelines(input_lines)
```

```python
filename = "result.txt"
Npad_top, Npad_bot, line_top, col_top, line_bot, col_bot = getINFO(filename)
```

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].hist(Npad_top, bins=64)
axs[0, 0].set_title('Histograma 1D: RPC superior', size=9)
axs[0, 0].set(ylabel="Contagem")
axs[1, 0].hist2d(col_top, line_top, bins=[8,8], cmap="YlOrRd", vmin=0, vmax=6000)
axs[1, 0].set_title('Histograma 2D: RPC superior', size=9)
axs[1, 0].set(ylabel="Coluna")
axs[0, 1].hist(Npad_bot, bins=64)
axs[0, 1].set_title('RPC inferior', size=9)
axs[1, 1].hist2d(col_bot, line_bot, bins=[8,8], cmap="YlOrRd", vmin=0, vmax=6000)
axs[1, 1].set_title('RPC inferior', size=9)
fig.suptitle("Contagem x PAD - run_1M_09_02_2023")
fig.savefig("contagem.jpg")

#for ax in axs.flat:
#    ax.set(xlabel='', ylabel='y-label')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=[7,3])
ax1.hist(Npad_top, bins=64)
ax2.hist(Npad_bot, bins=64)

ax1.set_title("RPC Superior")
ax2.set_title("RPC Inferior")
ax1.set_xlabel("Linha")
ax1.set_ylabel("Contagem")

plt.show()
```
